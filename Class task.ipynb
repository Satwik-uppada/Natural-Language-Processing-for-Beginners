```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def get_top_words(filepath, top_n=10):
    with open(filepath,'r',encoding='utf-8') as file:
        text = file.read()
        
    words = word_tokenize(text.lower())
    
    stop_words = set(stopwords.words('english'))  #
    filter_words = [word for word in words if word.isalnum() and word not in stop_words]
    word_freq = Counter(filter_words)
    
    top_words = word_freq.most_common(top_n)
    
    return top_words


file_path = "demo.txt"
top_words_in_file = get_top_words(file_path,top_n=10)

print(top_words_in_file)

    

```

```python
from nltk.corpus import wordnet

def word_similarity(word1, word2):
    synsets1 = wordnet.synsets(word1)
    synsets2 = wordnet.synsets(word2)
    
    max_similarity = 0
    
    for synset1 in synsets1:
        for synset2 in synsets2:
            similarity = synset1.wup_similarity(synset2)
            if similarity is not None and similarity > max_similarity:
                max_similarity = similarity
                
    return max_similarity
```

## **WORD MOVER DISTANCE**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import gensim.downloader as api
from gensim.models import Word2Vec

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python

def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text.lower())
    return [token for token in tokens if token.isalnum() and token not in stop_words]
```

```python

def word_mover_distance(text1, text2, model):
# Preprocess the texts
    tokens1 = preprocess_text(text1)
    tokens2 = preprocess_text(text2)

    # Compute Word Mover's Distance
    distance = model.wmdistance(tokens1, tokens2)
    return distance
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python

def extract_topic(texts, num_topics=5):
    tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
    
    tfidf = tfidf_vectorizer.fit_transform(texts)
    
    nmf_model = NMF(n_components=num_topics, random_state=42)
    nmf_model.fit(tfidf)   
    
    featuer_names = tfidf_vectorizer.get_feature_names_out()
    
    topics = []
    for topic_idx, topic in enumerate(nmf_model.components_):
        top_words_idx = topic.argsort()[-10:-1:-1] 
        top_words = [featuer_names[i] for i in top_words_idx]
        topics.append({"Topic":topic_idx,"Top words": top_words})
        
    return topics
```

```python
import nltk
from gensim.models import Word2Vec
import gensim.downloader as api
```

```python
nltk.download('punkt')
```

```python
sentences = [
    "The sun rises in the east.",
    "Birds chirp melodiously in the morning.",
    "The river flows gently through the valley.",
    "The stars twinkle in the night sky.",
    "Autumn leaves rustle in the wind.",
    "The city never sleeps, bustling with activity day and night.",
    "Waves crash against the shore, creating a soothing rhythm.",
    "The mountains stand majestically, touching the sky.",
    "Children play joyfully in the park.",
    "The aroma of freshly baked bread fills the air."
]
```

```python
tokenized_sentences = [nltk.word_tokenize(sentence.lower()) for sentence in sentences]

```

```python
dataset_name = 'text8'

dataset = api.load(dataset_name)

model = Word2Vec(sentences=dataset, vector_size=100, window=5, min_count=1 , workers=4)
model.save('word2vec_model')

word2vec_model = Word2Vec.load('word2vec_model')
```

```python
def get_word_vector(word):
    try:
        return word2vec_model.wv[word]
    except Exception:
        return None
```

```python
word_vectors = [[get_word_vector(word) for word in sentence] for sentence in tokenized_sentences]

for i, sentence_vectors in enumerate(word_vectors):
    print(f'\n snetence {i+1} word vectors:')
    for j, word_vector in enumerate(sentence_vectors):
        if word_vector is not None:
            print(f'\n wprd {j+1}: {word_vector}')
```
